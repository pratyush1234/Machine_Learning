```python
def fixpt(g,x0,Ea=1.e-7,maxit=30):
  """
  This function solves x = g(x) using fixed-point iteration.
  this method is repeated until either the relative error falls 
  below Ea (default 1.e-7) or reaches maxit (default 30).
  Input:
  g = name of the function for g(x)
  x0 = initial guess for x
  Ea = relative error thereshold 
  maxit = maximum number of iterations
  output:
  x1 = solution estimate
  ea = relative error
  i+1 = number of itrations
"""

  for i in range(maxit):
    x1 = g(x0)
    ea = abs((x1-x0)/x1)
    #x = np.zeros(maxit)
    #x[i+1] = g(x[i])
    #ea = abs((x[i+1]-x[i]/x[i+1]))
    if ea<Ea:
      break
    x0 = x1
    return x1,ea,i+1

# above was function defintion, below is calling the function
import numpy as np

def g(x):
  return np.sin(x)
  x0 = 0
  xsoln,ea,n = fixpt(g,x0,Ea=1.e-5)
  print('solution = ',xsoln)
  print('Relative error = ',ea)
  print('Number of iterations = ',n)



  # why print is not showing
"""
import numpy as np
rho = 1.23 # kg/m3
mu = 1.79e-5 # Pa*s
D = 0.005 # m
V = 40 # m/s
eps = 1.5e-6 # m
Re = rho*V*D/mu # Reynolds number
def g(f):
  fr =  np.log10(eps/3.7/D+1.26/Re/np.sqrt(f))
  # -4 has to be multiplied in the above expression
  return 1/fr**2
  f0 = 0.001
  fsoln,ea,n = fixpt(g,f0)
  print('Fanning friction factor =',fsoln)
  print('Relative error =',ea)
  print('Number of iterations =',n)
"""


       
       
```

```python
import numpy as np
def GaussSeidel(A,b,es=1.e-7,maxit=50):
  """
  the function implments gauss-seidel method to solve a
  set of linear algebraic equations without relaxation
  A = coefficient matrics 
  b = constant vector
  es = stopping criterion (default = 1.e-7)
  maxit = maximum number of iteration
  x = solution vector
  """
  # shape gives dimension of an array
  n,m = np.shape(A)
  print('n=',n,'m=',m) 
  if n != m :
      return 'coefficient of matrix must be square'
  c = np.zeros((n,n))
  #print('c=',c)
  x = np.zeros((n,1))
  #print('x=',x)
  # c with zeros in diagonal
  for i in range(n):
      for j in range(n): 
          if i != j:
              c[i,j] = A[i,j]
              print('c=',c)
  d = np.zeros((n,1))
  print('d=',d)
  for i in range(n):
      c[i,0:n] = c[i,0:n]/A[i,i] # divide c elements by A pivot
      print('co=',c)
      d[i] = b[i]/A[i,i]  # b array should be given
      print('do=',d)
  ea = np.zeros((n,1))
  xold = np.zeros((n,1))
  print('xold=',xold)
  for it in range(maxit):   # Gauss-Seidel
      for i in range(n):    # Gauss-Seidel method 
          xold[i] = x[i]  
          #print('xi=',x[i])  # save the x's for convergence test

      for i in range(n):
          x[i] = d[i] - c[i,:].dot(x)
          #print('xi=',x[i])
          k,l = np.shape(x)
          #print('k=',k,'l=',l)
          if x[i]!=0:
            # compute change error
              ea[i] = abs((x[i]-xold[i])/x[i])
              print('ea=',ea)
  if np.max(ea) < es: 
    return x 
    #break
  if it == maxit:
    return 'maximum iterations reached'
  else:
      return x

                        
A = np.array([[3,-0.1,-0.2],[0.1,7,-0.3],[0.3,-0.2,10]])
b = np.array([[7.85],[-19.3],[71.4]])
#print(A)
x = GaussSeidel(A,b,es=1.e-7,maxit=50)
print('x =',x)
                      
                  
```
