Import python libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn.metrics import mean_absolute_error
import joblib
```

import dataset

```python
from google.colab import files
uploaded = files.upload()
```

create a dataframe

```python
df = pd.read_csv('cardata.csv')
```

display some rowa to check data is uploaded or not

```python
df.head(n=5)
```

scrub the dataset

```python
del df['Transmission']
```

```python
del df['Owner']
```

```python
del df['Seller_Type']
```

```python
df.head(n=5)
```

one hot encoding to to convert the non-numerical values to numerical values and store in new dataframe

```python
features_df = pd.get_dummies(df, columns=['Fuel_Type'])
```

```python
features_df.head(n=11)
```

Delete dependent variable

```python
del features_df['Selling_Price']
```

convert dataframe into numpy array version to perform calculation

```python
X = features_df.values
y = df[Selling_Price].values
```

split the data into train-test pair in ratio of 70/30

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

select the supervised learning algorithm and chose its hyperparameters

```python
model = ensemble.GradientBoostingRegressor(
n_estimators=150,
learning_rate=0.1,
max_depth=30,
min_samples_split=4,
min_samples_leaf=6,
max_features=0.6,
loss='huber'
)
```

scikit learns fit function to train the model

```python
model.fit(X_train, y_train)
```

Scikit learns dump function to save the trained model as pkl file

```python
joblib.dump(model, 'car_trained_model.pkl')
```

Evaluate the result on training data

```python
mse = mean_absolute_error(y_train, model.predict(X_train))
print ("Training Set Mean Absolute Error: %.2f" % mse)
```

Evaluate the result on testing data

```python
mse = mean_absolute_error(y_test, model.predict(X_test))
print ("Test Set Mean Absolute Error: %.2f" % mse)
```
